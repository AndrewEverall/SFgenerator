# 1D markov plot

```python
# Plots for each of the 4 quantities
from seestar import ArrayMechanics

def HistogramValues(cat, label, nbin=20, rng=(0,1),
                   lweight='', weighted=False):   

    if  weighted: 
        density, val = np.histogram(cat[label], weights=cat[lweight], bins=nbin, range=rng)
        N = np.sum(cat[lweight])
    else: 
        density, val = np.histogram(cat[label], bins=nbin, range=rng)
        N = float(len(cat))
        
    centers = ArrayMechanics.BoundaryToCentre(val)
    
    return density/N, centers


# Plots for each of the 4 quantities

def CDFValues(cat, label, nbin=20, rng=(0,1),
                   lweight='', weighted=False):   

    if  weighted: 
        density, val = np.histogram(cat[label], weights=cat[lweight], bins=nbin, range=rng)
        N = np.sum(cat[lweight])
    else: 
        density, val = np.histogram(cat[label], bins=nbin, range=rng)
        N = float(len(cat))
        
    centers = ArrayMechanics.BoundaryToCentre(val)
    
    density = density/N
    cdf = np.zeros(len(density))
    for i in range(len(density)):
        cdf[i] = np.sum(density[:i])
    
    return cdf, centers
```

```python
plt.figure(figsize=(20, 20))

df_age = df1[['ageGyr', 'union']].copy()

N=10
for i in range(N):
    df_rand = df_age[np.random.rand(len(df_age)) < df_age.union].copy()
    
    age_d, age = HistogramValues(df_rand, 'ageGyr', nbin=40, rng=(0, 13))
    kwargs = {'linewidth':2., 'alpha':0.1, 'c':'b'}
    if i==0: kwargs['label']='reweighted'
    plt.plot(age, age_d, **kwargs)#label='reweighted', linewidth=2., alpha=0.1, c='b')
    
age_d, age = HistogramValues(df1, 'ageGyr', nbin=40, rng=(0, 13))
plt.plot(age, age_d, label='fulldf', linewidth=2., alpha=1., c='r')

age_d, age = HistogramValues(survey, 'ageGyr', nbin=40, rng=(0, 13))
plt.plot(age, age_d, label='survey', linewidth=2., alpha=1., c='g')

plt.legend()
```

```python
def kde(df, survey, coord, ax, xlabel={}, xlim={}, fs=50, legend=True,
        label1=r'$f_\mathrm{DF}(\mathbf{v} \,|\,\Theta_1)$',
        label2=r'$f_\mathrm{Obs}(\mathbf{v} \,|\,\Theta_1)$',
        label3=r'$f_\mathrm{DF}(\mathbf{v} \,|\,\Theta_1) \times \mathrm{P}(\mathrm{S}\,|\,\mathbf{v}, \Theta_1)$'
       ):


    
    plt.sca(ax)

    sns.kdeplot(df[coord], label=label1, alpha=1., c='r', legend=legend, linestyle='-.', linewidth=2)
    sns.kdeplot(survey[coord], label=label2, alpha=1., c='g', legend=legend,linestyle='-', linewidth=3)

    N=10
    for i in range(N):
        df_rand = df[np.random.rand(len(df)) < df.union][[coord]].copy()

        kwargs = {'alpha':0.2, 'color':'b'}
        if i==0: labelargs = {'label':label3,\
                              'legend':legend}
        else: labelargs = {'legend':False}
        sns.kdeplot(df_rand[coord], **kwargs, **labelargs, linestyle='--', linewidth=2)

    plt.xlim(xlim[coord])
    plt.xlabel(xlabel[coord], fontsize=fs)
    plt.ylabel(ylabel[coord], fontsize=fs)
    if legend: plt.setp(ax.get_legend().get_texts(), fontsize=fs) 

    plt.tick_params('both', labelsize=20)
```
